```matlab
gsxyz = xlsread('gs\gs-1429_n_AAV.xlsx');
taxyz = xlsread('ta\ta_1653_n_djm.xlsx');
```

```matlab
gckgs = gckernel();
gckta = gckernel();
setKernel(gckgs,[3,3]); setKernel(gckta,[3,3])
gckgs.setData(gsxyz(:,1),[0.02,0]); %this set smoothness of kernel in units of data. If omitted, it reverts to Bhumbra & Dyball
gckta.setData(taxyz(:,1),[0.02,0]); % Change first number to test
gckgs.convolve(gsxyz(:,1));
gckta.convolve(taxyz(:,1));
```

```matlab

fh = figure;
set(fh, 'Name', '1D Gaussian convolution example')

subplot(1, 4, 1);
plot(gckgs.cx);
axis tight;
%axis off;
title('Gaussian kernel');

subplot(1, 4, 2);
hist(gsxyz(:,1), 20);
ax = gca;
axis tight;
title('GS histogram');

subplot(1, 4, 3);
hist(taxyz(:,1), 20);
ax = gca;
axis tight;
title('TA histogram', 'Color','r');

subplot(1, 4, 4);
plgs = plot(gckgs.x, gckgs.p);
hold on
plta = plot(gckta.x,gckta.p);
set(plta,'Color','r');
xlim(ax.XLim);
%axis tight;
title('Convolved PMF');



```
