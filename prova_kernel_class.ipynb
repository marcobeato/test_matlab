##prova kernel with clases


```matlab
x = [normrnd(10, 10, 1, 1000),normrnd(30, 5, 1, 1000),normrnd(50, 10, 1, 1000)];
gsxyz = readmatrix('gs\gs-1429_n_AAV.xlsx');
taxyz = readmatrix('ta\ta_1653_n_djm.xlsx');
```

```matlab
gck1 = gckernel();
setKernel(gck1,[3,3]);
gck1.setData(x,[0.01,0]); %this set smoothness of kernel in units of data. If omitted, it reverts to Bhumbra & Dyball
gck1.convolve(xyz(:,1));
```

```matlab
fh = figure;
set(fh, 'Name', '1D Gaussian convolution example')

subplot(1, 3, 1);
plot(gck1.cx);
axis tight;
%axis off;
title('Gaussian kernel');

subplot(1, 3, 2);
hist(xyz(:,1), 20);
ax = gca;
axis tight;
title('Coarse histogram');

subplot(1, 3, 3);
plot(gck1.x, gck1.p);
xlim(ax.XLim);
%axis tight;
title('Convolved PMF');



```
